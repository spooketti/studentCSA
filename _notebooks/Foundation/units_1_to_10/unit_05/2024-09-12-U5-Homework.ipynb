<script type="module">
    import GameControl from '{{site.baseurl}}/assets/js/nitd_rpg/GameControl.js';
    import Game from '{{site.baseurl}}/assets/js/nitd_rpg/Player.js';  // Import the Game class

    // Create an instance of the game
    const gameInstance = new Game();

    // Background data
    const image_src = "{{site.baseurl}}/images/nitd_rpg/maze3.png";
    const image_data = {
        pixels: {height: 580, width: 1038}
    };
    const image = {src: image_src, data: image_data};

    // Sprite data
    const sprite_src = "{{site.baseurl}}/images/rpg/turtle.png";
    const sprite_data = {
        SCALE_FACTOR: 15,
        STEP_FACTOR: 1000,
        ANIMATION_RATE: 50,
        pixels: {height: 280, width: 256},
        orientation: {rows: 4, columns: 3 },
        down: {row: 0, start: 0, columns: 3 },
        left: {row: 1, start: 0, columns: 3 },
        right: {row: 2, start: 0, columns: 3 },
        up: {row: 3, start: 0, columns: 3 },
    };
    const sprite = {src: sprite_src, data: sprite_data};

    // Assets for game
    const assets = {image: image, sprite: sprite};

    // Start game engine
    GameControl.start(assets);

    // Timer logic
    let startTime = Date.now();
    let timerRunning = true;

    function updateTimer() {
        if (gameInstance.isMazeCompleted()) {  // Check maze completion via the method
            timerRunning = false;  // Stop the timer
        }

        if (timerRunning) {
            const currentTime = Date.now();
            const elapsedSeconds = Math.floor((currentTime - startTime) / 1000);
            document.getElementById('timer').innerText = `Time: ${elapsedSeconds}s`;
            requestAnimationFrame(updateTimer);
        }
    }

    // Start the timer and game mechanics
    window.onload = () => {
        updateTimer();
        moveImage();  // Start the image moving
    };
</script>

<script>
    // Variables to control image movement
    let posX = Math.random() * (document.documentElement.scrollWidth - 50);
    let posY = Math.random() * (document.documentElement.scrollHeight - 50);
    let velocityX = 5;
    let velocityY = 5;

    // Function to move the image
    function moveImage() {
        const image = document.getElementById('randomImage');
        const imageWidth = image.offsetWidth;
        const imageHeight = image.offsetHeight;

        // Update the position
        posX += velocityX;
        posY += velocityY;

        // Bounce off the walls
        if (posX <= 0 || posX + imageWidth >= document.documentElement.scrollWidth) {
            velocityX = -velocityX;
        }
        if (posY <= 0 || posY + imageHeight >= document.documentElement.scrollHeight) {
            velocityY = -velocityY;
        }

        // Apply the new position
        image.style.left = `${posX}px`;
        image.style.top = `${posY}px`;

        // Call the moveImage function repeatedly to keep the image moving
        requestAnimationFrame(moveImage);
    }
</script>
